```intel
mov ecx,2004
mov eax,ecx
mov edx,0
mov ebx,4
idiv ebx
cmp edx,0  ; se revisa que sea multiplo de 4
    je hundred
    mov ecx,0
    jmp ends
hundred: mov eax,ecx ;si es multiplo de 4 se revisa que sea multiplo de 100
         mov edx,0
         mov ebx,100
         idiv ebx
         cmp edx,0
             je four
             mov ecx,1
             jmp ends
four: mov eax,ecx ; si es multiplo de 100 se revisa que sea multiplo de 400
         mov edx,0
         mov ebx,400
         idiv ebx
         cmp edx,0
             je sies
             mov ecx,0
             jmp ends
sies: mov ecx,1

ends: mov ecx, ecx ;si ecx es 1 es leap year, si es 0 no es leap year 


    


```
